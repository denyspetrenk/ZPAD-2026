## Підготовка середовища

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal import butter, filtfilt
import ipywidgets as widgets
from IPython.display import display

sns.set_theme(style="whitegrid")
```

## Логіка та інтерфейс

```python
def plot_harmonic(amp, freq, phase, noise_mean, noise_cov, cutoff, show_noise):
    t = np.linspace(0, 10, 1000)

    y_clean = amp * np.sin(2 * np.pi * freq * t + phase)

    # Шум
    std_dev = np.sqrt(noise_cov) if noise_cov > 0 else 0

    np.random.seed(42)
    noise = np.random.normal(noise_mean, std_dev, len(t))
    y_noisy = y_clean + noise

    nyquist = 0.5 * (1.0 / (t[1] - t[0]))
    normal_cutoff = cutoff / nyquist if (cutoff / nyquist) < 1.0 else 0.99
    b, a = butter(4, normal_cutoff, btype='low', analog=False)
    y_filtered = filtfilt(b, a, y_noisy)

    plt.figure(figsize=(12, 6))
    if show_noise:
        plt.plot(t, y_noisy, color='orange', alpha=0.5, label='Зашумлена')
    plt.plot(t, y_clean, color='blue', linestyle='--', linewidth=2, label='Чиста')
    plt.plot(t, y_filtered, color='purple', linewidth=2, label='Відфільтрована')

    plt.title("Гармоніка")
    plt.ylim(-3, 3)
    plt.legend(loc='upper right')
    plt.show()
```

## Створення віджету для Colab

```python
widgets.interact(plot_harmonic,
                 amp=widgets.FloatSlider(value=1.0, min=0.1, max=2.0, step=0.1, description='Амплітуда'),
                 freq=widgets.FloatSlider(value=0.25, min=0.01, max=2.0, step=0.01, description='Частота'),
                 phase=widgets.FloatSlider(value=0.0, min=0, max=2*np.pi, step=0.1, description='Фаза'),
                 noise_mean=widgets.FloatSlider(value=0.0, min=-1.0, max=1.0, step=0.1, description='Шум (Середнє)'),
                 noise_cov=widgets.FloatSlider(value=0.1, min=0.0, max=1.0, step=0.05, description='Шум'),
                 cutoff=widgets.FloatSlider(value=1.5, min=0.1, max=5.0, step=0.1, description='Зріз фільтра'),
                 show_noise=widgets.Checkbox(value=True, description='Показати шум'))
```
